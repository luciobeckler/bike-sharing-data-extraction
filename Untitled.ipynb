```python
import numpy as np
import pandas as pd


df = pd.read_csv("https://pycourse.s3.amazonaws.com/bike-sharing.csv")

#INFORMAÇÕES SOBRE O DATASET:
df["datetime"]=pd.to_datetime(df["datetime"])

#print(df.dtypes,df.info,df.describe)

def findSeasonIDMin(df):
    seasonMinCount=float ('inf')
    for i in range(1, 5):
        seasonCount = df['total_count'].loc[df['season']==i].mean()
        if(seasonCount<seasonMinCount):
            seasonMinCount = seasonCount
            seasonID = i
    if(seasonID==1): seasonID = 'inverno'
    elif(seasonID==2): seasonID = 'primavera'
    elif(seasonID==3): seasonID = 'verão'
    elif(seasonID==4): seasonID = 'outono'
    return(seasonID)



def findSeasonIDMax(df):
    seasonMaxCount=0
    for i in range(1, 5):
        seasonCount = df['total_count'].loc[df['season']==i].mean()
        if(seasonCount>seasonMaxCount):
            seasonMaxCount = seasonCount
            seasonID = i
    if(seasonID==1): seasonID = 'inverno'
    elif(seasonID==2): seasonID = 'primavera'
    elif(seasonID==3): seasonID = 'verão'
    elif(seasonID==4): seasonID = 'outono'
    return(seasonID)



def findMoreRequiredHour(df):
    aux = 0
    for i in range(0,24):
        if(df['total_count'].loc[df['hour']==i].mean()>aux): 
            aux = df['total_count'].loc[df['hour']==i].mean()
            moreRequiredHour = i
    return(moreRequiredHour)



def findLessRequiredHour(df):
    aux = float('inf')
    for i in range(0,24):
        if(df['total_count'].loc[df['hour']==i].mean()<aux): 
            aux = df['total_count'].loc[df['hour']==i].mean()
            lessRequiredHour = i
    return(lessRequiredHour)



def findMoreRequiredWeekday(df):
    aux = 0
    for i in range(0,7):
        if(df['total_count'].loc[df['weekday']==i].mean() > aux):
            aux = df['total_count'].loc[df['weekday']==i].mean()
            moreRequiredWeekday = i
    return (findWeekdayByID(moreRequiredWeekday))



def findLessRequiredWeekday(df):
    aux = float('inf') 
    for i in range (0,7):
        if(df['total_count'].loc[df['weekday']==i].mean() < aux):
            aux = df['total_count'].loc[df['weekday']==i].mean()
            lessRequiredWeekday = i
    return (findWeekdayByID(lessRequiredWeekday))


  
def findMoreRequiredHourInAWeekday(df,day):
    aux = 0
    for i in range(0,24):
        if(df['total_count'].loc[df['weekday']==day].loc[df['hour']==i].mean() > aux):
            aux = df['total_count'].loc[df['weekday']==day].loc[df['hour']==i].mean()
            moreRequiredHourInAWeekday = i
    return(moreRequiredHourInAWeekday)


    
def findWeekdayByID(day):
    if(day == 0): day = "domingo"
    elif(day == 1): day = "segunda-feira"
    elif(day == 2): day = "terça-feira"
    elif(day == 3): day = "quarta-feira"
    elif(day == 4): day = "quinta-feira"
    elif(day == 5): day = "sexta-feira"
    elif(day == 6): day = "sábado"
    
    return (day)

            
            
print("a. Qual o tamanho desse dataset?", "O dataset possui", df.shape[0],"linhas e ",df.shape[1],"colunas")

print("b. Qual a média da coluna windspeed?", "A média da coluna windspeed é:", df["windspeed"].mean())

print("c. Qual a média da coluna temp?", "A média da coluna temp é:", df["temp"].mean())

print("d. Quantos registros existem para o ano de 2011?", "No ano de 2011 ocorreram ",df['registered'].loc[df['datetime']<'2012-01-01'].sum(), "registros de locações de bicicletas" )

print("e. Quantos registros existem para o ano de 2012?", "No ano de 2011 ocorreram ",df['registered'].loc[(df["datetime"]>="2012-01-01") & (df["datetime"] < "2013-01-01")].sum(), "registros de locações de bicicletas")

print("f. Quantas locações de bicicletas foram efetuadas em 2011?","No ano de 2011 ocorreram ",df['total_count'].loc[df['datetime']<'2012-01-01'].sum(),"locações de bicicletas")

print("g. Quantas locações de bicicletas foram efetuadas em 2012?","No ano de 2012 ocorreram", df['total_count'].loc[(df['datetime']>='2012-01-01')& (df['datetime']<'2013-01-01')].sum(), "locações de bicicletas")

print ("h. Qual estação do ano contém a maior média de locações de bicicletas?", 'A estação do ano que possui a maior média de locação de bikes é o/a',findSeasonIDMax(df))

print('i. Qual estação do ano contém a menor média de locações de bicicletas?','A estação do ano que possui a menor média de locação de bikes é o/a',findSeasonIDMin(df))

print('j. Qual horário do dia contém a maior média de locações de bicicletas?','A hora que possui a maior média de locações de bike são ás',findMoreRequiredHour(df),'horas')

print('k. Qual horário do dia contém a menor média de locações de bicicletas?','A hora que possui a maior média de locações de bike são ás',findLessRequiredHour(df),'horas')

print('l. Que dia da semana contém a maior média de locações de bicicletas?','O dia da semana que possui a maior média de locações de bike é',findMoreRequiredWeekday(df),)

print ('m. Que dia da semana contém a menor média de locações de bicicletas?', 'O dia da semana que possui a menor média de locações de bike é',findLessRequiredWeekday(df))

print('n. Às quartas-feiras (weekday = 3), qual o horário do dia contém a maior média de locações de bicicletas?', 'O horário que possui a maior média de locações de bikes em uma quarta são às',findMoreRequiredHourInAWeekday(df,3),'horas')

print('n. Às quartas-feiras (weekday = 3), qual o horário do dia contém a maior média de locações de bicicletas?', 'O horário que possui a maior média de locações de bikes em um sábado são às',findMoreRequiredHourInAWeekday(df,6),'horas')
```
